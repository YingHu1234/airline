```python
import pandas as pd
df=pd.read_csv('airline review.csv')
df
```

```python
df.recommended[df.recommended == 0] = 'neg'
df.recommended[df.recommended == 1] = 'pos'
df
```

# Negatvie reviews:

```python
df_neg=df.loc[df['recommended']=='neg']
df_neg
```

```python
import nltk
import gensim
import re
from gensim import corpora
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
stopword=stopwords.words('english')



df1_neg=df_neg.loc[df_neg['airline_name']=="air-canada-rouge",['content']]
df2_neg=df_neg.loc[df_neg['airline_name']=="british-airways",['content']]
df3_neg=df_neg.loc[df_neg['airline_name']=="united-airlines",['content']]



list_all_neg=[df1_neg['content'].tolist(),df2_neg['content'].tolist(),df3_neg['content'].tolist()]
# print(list_all_neg)
print(len(list_all_neg))
```

```python


# list_all_neg=[str(df1_neg['content'].tolist()).strip('[]'),str(df2_neg['content'].tolist()).strip('[]'),str(df3_neg['content'].tolist()).strip('[]')]
# # print(list_all_neg)
# print(list_all_neg)



def clean(text):
    
    filtered=[x.lower() for x in nltk.word_tokenize(text) if x.lower() not in stopword and x.isalpha()]   
    frequency=nltk.FreqDist(filtered)
    stem=[stemmer.stem(x) for x in filtered]

    
    return stem


cleaned1=[clean(x) for x in list_all_neg[0]]
cleaned2=[clean(x) for x in list_all_neg[1]]
cleaned3=[clean(x) for x in list_all_neg[2]]


print(cleaned1)





```

# air-canada-rouge airline:

```python
dictionary=corpora.Dictionary(cleaned1)

for i in dictionary.keys():
    print(i,dictionary[i])


matrix1=[dictionary.doc2bow(x) for x in cleaned1]
print(matrix1)
print(len(dictionary))
```

```python
# View

[[(dictionary[id], freq) for id, freq in cp] for cp in matrix1[:1]]
```

```python
#corpus_tfidf:

from gensim import corpora, models
tfidf = models.TfidfModel(matrix1)
corpus_tfidf = tfidf[matrix1]
from pprint import pprint
for doc in corpus_tfidf:
    pprint(doc)
    break
```

```python
#setup lda model:
lda=gensim.models.ldamodel.LdaModel
model=lda(matrix1, num_topics=10, id2word=dictionary)
print(model.print_topics(num_topics=10,num_words=40))
```

```python
#get the weight of the words for first doc:

weight=re.findall('0\.[0-9]{3}', model.print_topics(num_topics=10, num_words=40)[0][1])
words=re.findall('"(.*?)"', model.print_topics(num_topics=10, num_words=40)[0][1])
topic=[(x,y) for x,y in zip(words, weight)]
topic_summary=dict(topic)
print(topic_summary)
```

```python
#get the topic distribution:

topic_distribution=[]
for i in range(len(cleaned1)):
    dist=model.get_document_topics(matrix1[i])
    empty=dict.fromkeys(range(10),0)
    empty.update(dist)
    topic_distribution.append(empty)
print(topic_distribution)
```

```python
#hhi:

def hhi(lst):
    return sum([(x*100)**2 for x in lst])
print([hhi(x.values()) for x in topic_distribution])
```

```python
#list of topic_distribution:

import gensim
Lda=gensim.models.ldamodel.LdaModel
ldamodel=lda(matrix1, num_topics=10, id2word=dictionary)

topic_distribution_list1=[]
for i in range(len(matrix1)):
    distribution=dict(ldamodel.get_document_topics(matrix1[i]))
    template_dict=dict.fromkeys(range(10),0)
    template_dict.update(distribution)
    topic_distribution_list1.append(template_dict)

print(topic_distribution_list1)
```

```python
def cossinesimilarity(x,y):
    numerator=sum([i*j for i,j in zip(x,y)])
    length_x=sum([k**2 for k in x])**0.5
    length_y=sum([k**2 for k in y])**0.5
    denominator=length_x*length_y
    return numerator/denominator
```

```python
#similarity:


for i in range(1, len(topic_distribution_list1)):
    topic_dist_doci=topic_distribution_list1[i].values()
    topic_dist_doc0=topic_distribution_list1[0].values()
    cs=cossinesimilarity(topic_dist_doci, topic_dist_doc0)
    print('the similarity between the {}th doc and the 1s doc is{}'. format(i,cs))
```
